## Lendo os dados do arquivo csv

```python
# Leitura dos dados
import csv
dadosRaw=[]
path = r"C:\Users\R211315\sinaisvitais003 100dias DV2 RAxxx5.txt"
with open(path,'r',newline='') as ARQUIVO:
    d = csv.reader(ARQUIVO)
    dd=list(d)
    for i in range(0,len(dd)):
        p=dd[i][0]  
        palavras=p.split("\t")
        dadosRaw.append({"HORA":palavras[0],
                      "BATIMENTO":palavras[1],
                      "PRESSAO":palavras[2],
                      "TEMPERATURA":palavras[3]}
                    )
```

## Estratgia 2 - Substituindo os valores dos outliers

```python
# Fazendo uma cpia dos dados brutos
dadosRawDF2 = dadosRawDF
```

### Rotina de ajuste dos outliers

```python
# Rotina que ajusta o valor de um ponto, em um determinado campo, pela mdia do anterior e do posterior
# reg = registro a ser alterado
# campo = coluna do DF
# tamanho = nmero de registros
# o cdigo no tratar de 2 campos seguidos com outliers

def media_outliers(reg,campo,tamanho):
   
  if reg == 0:   # primeiro registro
    dadosRawDF2.loc[0,campo] = dadosRawDF2.loc[1,campo]
  else:
    if reg == tamanho:
        dadosRawDF2.loc[reg,campo] = dadosRawDF2.loc[reg-1,campo]
    else:
        dadosRawDF2.loc[reg,campo] = (((dadosRawDF2.loc[reg-1,campo])+(dadosRawDF2.loc[reg+1,campo]))/2)
```

```python
# Vamos utilizar os vetores que j possuem os registros com os dados com os outliers
print('Registros de batimento com outliers:',indicesOutliersBatimento)
print('Registros de presso com outliers:',indicesOutliersPressao)
print('Registros de temperatura com outliers:',indicesOutliersTemperatura)
```

```python
# Ajuste dos outliers dos batimentos

print(f'Batimentos antes da alterao: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', 
      dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',
      dadosRawDF2.loc[1490,'BATIMENTO'])

for i in range(0,len(indicesOutliersBatimento)):
   media_outliers(indicesOutliersBatimento[i],'BATIMENTO',len(dadosRawDF2))

print(f'Batimentos aps da alterao: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', 
      dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',
      dadosRawDF2.loc[1490,'BATIMENTO'])
```

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
# Variveis preditoras (independentes) - batimentos e presso
X=mediaPorDia2.drop(['TEMPERATURA'],axis=1)
# Varivel dependente - temperatura
y=mediaPorDia2.drop(['BATIMENTO','PRESSAO'],axis=1)
```

```python
# Criando um modelo de regresso linear
modelo=LinearRegression()
# Terminando o modelo
modelo.fit(X,y)
```

```python
# Coeficiente de determinao
modelo.score(X,y)
```

A interpretao da sada acima  que, em mdia, as variveis preditoras PRESSAO e BATIMENTO, explicam 90,1% das variaes da varivel dependente TEMPERATURA.

```python
# Intercepto (constante)
modelo.intercept_
```

```python
# Coeficientes
modelo.coef_
```

Resposta: O modelo acima  permite determinar a temperatura a partir dos dados da presso e batimentos com 90,1 % de certeza.

```python
# Moda por dia
from statistics import mode
dadosModaPorDia= dadosRawDFIndexDia2.groupby(['DIA'])
def calculaModa(dia,coluna):
    return mode(dadosModaPorDia.get_group(dia)[coluna])

batimento=[]
pressao=[]
temperatura=[]
for name, group in dadosModaPorDia:
    batimento.append(calculaModa(name,'BATIMENTO'))
    pressao.append(calculaModa(name,'PRESSAO'))
    temperatura.append(calculaModa(name,'TEMPERATURA'))
dia=[]
for i in range(1,101):
    dia.append(i)
    
modaPorDia=pd.DataFrame(list(zip(batimento,pressao,temperatura,dia)),columns=['BATIMENTO',
                                                                         'PRESSAO', 'TEMPERATURA','DIA'])
modaPorDia=modaPorDia.set_index('DIA')
```
