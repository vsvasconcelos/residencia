```python
# Leitura dos dados
import csv
dadosRaw=[]
path = r"C:\Users\R211315\sinaisvitais003 100dias DV2 RAxxx5.txt"
with open(path,'r',newline='') as ARQUIVO:
    d = csv.reader(ARQUIVO)
    dd=list(d)
    for i in range(0,len(dd)):
        p=dd[i][0]  
        palavras=p.split("\t")
        dadosRaw.append({"HORA":palavras[0],
                      "BATIMENTO":palavras[1],
                      "PRESSAO":palavras[2],
                      "TEMPERATURA":palavras[3]}
                    )
```

## Estratégia 2 - Substituindo os valores dos outliers

```python
# Fazendo uma cópia dos dados brutos
dadosRawDF2 = dadosRawDF
```

### Rotina de ajuste dos outliers

```python
# Rotina que ajusta o valor de um ponto, em um determinado campo, pela média do anterior e do posterior
# reg = registro a ser alterado
# campo = coluna do DF
# tamanho = número de registros
# o código não tratará de 2 campos seguidos com outliers

def media_outliers(reg,campo,tamanho):
   
  if reg == 0:   # primeiro registro
    dadosRawDF2.loc[0,campo] = dadosRawDF2.loc[1,campo]
  else:
    if reg == tamanho:
        dadosRawDF2.loc[reg,campo] = dadosRawDF2.loc[reg-1,campo]
    else:
        dadosRawDF2.loc[reg,campo] = (((dadosRawDF2.loc[reg-1,campo])+(dadosRawDF2.loc[reg+1,campo]))/2)
```

```python
# Vamos utilizar os vetores que já possuem os registros com os dados com os outliers
print('Registros de batimento com outliers:',indicesOutliersBatimento)
print('Registros de pressão com outliers:',indicesOutliersPressao)
print('Registros de temperatura com outliers:',indicesOutliersTemperatura)
```

```python
# Ajuste dos outliers dos batimentos

print(f'Batimentos antes da alteração: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', 
      dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',
      dadosRawDF2.loc[1490,'BATIMENTO'])

for i in range(0,len(indicesOutliersBatimento)):
   media_outliers(indicesOutliersBatimento[i],'BATIMENTO',len(dadosRawDF2))

print(f'Batimentos após da alteração: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', 
      dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',
      dadosRawDF2.loc[1490,'BATIMENTO'])
```

Para tentar responder esta questão utilizaremos a medida estatística Coeficiente de Variação:
> Em teoria das probabilidades e estatística, o coeficiente de variação (CV), também conhecido como desvio padrão relativo (DPR), é uma medida padronizada de dispersão de uma distribuição de probabilidade ou de uma distribuição de frequências. É frequentemente expresso como uma porcentagem, sendo definido como a razão do desvio padrão σ pela média μ (ou seu valor absoluto, | μ |.   
[Fonte](https://pt.wikipedia.org/wiki/Coeficiente_de_varia%C3%A7%C3%A3o)

### Coeficiente de variação

```python
# Calculo do coeficiente de variação
coeficienteDeVariacao=(desvioPadraoPorDia2/mediaPorDia2)*100
```

```python
# Impressão do coeficiente de variação diário
coeficienteDeVariacao.plot(y=['BATIMENTO','PRESSAO','TEMPERATURA'],use_index=True)
```

Conforme o gráfico, a temperatura possui um CV menor que 1% e praticamente constante.   
Já o batimento, tem um CV por volta de 4%, contudo ele oscila um pouco mais que a temperatura.   
Por sua vez, a temperatura possui o maior CV, entre 9 e 12%.   
Desta forma, parece que a utilização das médias representam bem a população, permitindo seu uso no cálculo das correlações.

Tentaremos prever a temperatura corporal em função da pressão e dos batimentos cardíacos (análise bivariada).

```python
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
# Variáveis preditoras (independentes) - batimentos e pressão
X=mediaPorDia2.drop(['TEMPERATURA'],axis=1)
# Variável dependente - temperatura
y=mediaPorDia2.drop(['BATIMENTO','PRESSAO'],axis=1)
```

```python
# Criando um modelo de regressão linear
modelo=LinearRegression()
# Terminando o modelo
modelo.fit(X,y)
```

```python
# Coeficiente de determinação
modelo.score(X,y)
```

A interpretação da saída acima é que, em média, as variáveis preditoras PRESSAO e BATIMENTO, explicam 90,1% das variações da variável dependente TEMPERATURA.

```python
# Intercepto (constante)
modelo.intercept_
```

```python
# Coeficientes
modelo.coef_
```

Resposta: O modelo acima é permite determinar a temperatura a partir dos dados da pressão e batimentos com 90,1 % de certeza.

## Moda

```python
# Moda por dia
from statistics import mode
dadosModaPorDia= dadosRawDFIndexDia2.groupby(['DIA'])
def calculaModa(dia,coluna):
    return mode(dadosModaPorDia.get_group(dia)[coluna])

batimento=[]
pressao=[]
temperatura=[]
for name, group in dadosModaPorDia:
    batimento.append(calculaModa(name,'BATIMENTO'))
    pressao.append(calculaModa(name,'PRESSAO'))
    temperatura.append(calculaModa(name,'TEMPERATURA'))
dia=[]
for i in range(1,101):
    dia.append(i)
    
modaPorDia=pd.DataFrame(list(zip(batimento,pressao,temperatura,dia)),columns=['BATIMENTO',
                                                                         'PRESSAO', 'TEMPERATURA','DIA'])
modaPorDia=modaPorDia.set_index('DIA')
```

```python
modaPorDia.plot(y=['BATIMENTO'],use_index=True,title='Moda dos batimentos cardiacos por dia')
```

```python
modaPorDia.plot(y=['PRESSAO'],use_index=True,title='Moda da pressão por dia')
```

```python
modaPorDia.plot(y=['TEMPERATURA'],use_index=True,title='Moda da temperatura por dia')
```
