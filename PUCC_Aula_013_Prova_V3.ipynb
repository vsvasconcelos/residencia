```python
# Leitura dos dados
import csv
dadosRaw=[]
path = "C:\\Users\\madad\Documents\\PUC_Campinas\\Aula 13 - Avaliação\sinaisvitais003 100dias DV2 RA5.txt"
with open(path,'r',newline='') as ARQUIVO:
    d = csv.reader(ARQUIVO)
    dd=list(d)
    for i in range(0,len(dd)):
        p=dd[i][0]  
        palavras=p.split("\t")
        dadosRaw.append({"HORA":palavras[0],
                      "BATIMENTO":palavras[1],
                      "PRESSAO":palavras[2],
                      "TEMPERATURA":palavras[3]}
                    )
```

```python
# Rotina que ajusta o valor de um ponto, em um determinado campo pela média do anterior e do posterior
# reg = registro a ser alterado
# campo = coluna do DF
# tamnanho = número de registros
# o código não tratará de 2 campos seguidos com outliers

def media_outliers(reg,campo,tamanho):
   
  if reg == 0:   # primeiro registro
    dadosRawDF2.loc[0,campo] = dadosRawDF2.loc[1,campo]
  else:
    if reg == tamanho:
        dadosRawDF2.loc[reg,campo] = dadosRawDF2.loc[reg-1,campo]
    else:
        dadosRawDF2.loc[reg,campo] = (((dadosRawDF2.loc[reg-1,campo])+(dadosRawDF2.loc[reg+1,campo]))/2)
```

```python
# Vamos utilizar os vetores que já possuem os registros com os dados com os outliers
print('Registros de batimento com outliers:',indicesOutliersBatimento)
print('Registros de pressão com outliers:',indicesOutliersPressao)
print('Registros de temperatura com outliers:',indicesOutliersTemperatura)
```

```python
# Ajuste dos outliers dos batimentos

print(f'Batimentos antes da alteração: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',dadosRawDF2.loc[1490,'BATIMENTO'])

for i in range(0,len(indicesOutliersBatimento)):
   media_outliers(indicesOutliersBatimento[i],'BATIMENTO',len(dadosRawDF2))

print(f'Batimentos após da alteração: ', dadosRawDF2.loc[203,'BATIMENTO'],' - ', dadosRawDF2.loc[600,'BATIMENTO'],' - ',dadosRawDF2.loc[1065,'BATIMENTO'],' - ',dadosRawDF2.loc[1490,'BATIMENTO'])
```
